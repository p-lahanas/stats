# Central Limit Theorem

Why do so many things seem to follow some sort of a normal distribution? Well, one possible explanation is the Central Limit Theorem.

**Central Limit Theorem:** \
"The sampling distribution of a sample mean is approximately normal if the sample size is large enough."

So what does this mean? If we calculate a sample mean over a distribution (regardless of the distribution), the distribution of sample means will approach a normal distribution.

```python
import numpy as np
import seaborn as sns

# Generate some sample means from a variety of distribution

def generate_sample_mean_uniform(num_samples:int =100) -> np.ndarray:
    return np.random.randint(1,15, size=num_samples).mean()

def generate_sample_means_uniform(samples: int = 100) -> np.ndarray:
    return np.array([generate_sample_mean_uniform() for _ in range(samples)])


sns.histplot(generate_sample_means_uniform())
```
